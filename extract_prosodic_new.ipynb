```python
import  numpy as np
import librosa
from numpy import mean, diff
from matplotlib.mlab import find
import os
import pandas as pd
import pickle
```

```python
# Function to extract the log energy per frames
def get_log_energy(sample_data):

    hann_win = np.hanning(len(sample_data))
    energy = np.log10(np.sum(np.square(sample_data * hann_win)))

    return energy


# Function to calculate the pitch based on
# zero rate. This function is taken as it is from
# https://gist.github.com/endolith/255291.
# We are thankful to the author for proposing this
# measure of pitch.
def freq_from_crossings(sig, fs):
    """
    Estimate frequency by counting zero crossings
    """
    # Find all indices right before a rising-edge zero crossing
    indices = find((sig[1:] >= 0) & (sig[:-1] < 0))

    # More accurate, using linear interpolation to find intersample
    # zero-crossings (Measures 1000.000129 Hz for 1000 Hz, for instance)
    crossings = [i - sig[i] / (sig[i+1] - sig[i]) for i in indices]

    # Some other interpolation based on neighboring points might be better.
    # Spline, cubic, whatever

    return fs / mean(diff(crossings))


# Function to calculate  the zero_rate crossing per frame
def get_zero_rate_crossing(sample):

    # Return the weighted average of the zero crossing rate
    return librosa.feature.zero_crossing_rate(sample)


def freq_from_fft(sig, fs):
    w = np.fft.fft(sig)
    freqs = np.fft.fftfreq(len(w))
    idx = np.argmax(np.abs(w))
    freq = freqs[idx]
    freq_in_hertz = abs(freq * fs)
    return (freq_in_hertz)
```

```python
log_energy = []
zero_cross_rate = []
pitch = []

path = "/Users/schmuck/Documents/Box Sync/Courses/FALL2018/MLSP/MLSP_Project/programs/Audio_Speech_Actors_01-24"

# get the pitch data.
pitch_data = pickle.load(open("/Users/schmuck/Documents/Box Sync/Courses/FALL2018/MLSP/MLSP_Project/programs/pitch_data","rb"))

for each in os.listdir(path):
    print (" #### Reading category ", each, " ##### ")
    path1 = path + '/' + each

    for each_song in os.listdir(path1):
        path2 = path1 + '/' + each_song
        print(path2)
        print ("Reading file ", each_song)

        data, sr = librosa.load(path2, sr=16000)
        log_energy.append(get_log_energy(data))
        #pitch.append(freq_from_crossings(data, sr))
        zero_cross_rate.append(np.mean(get_zero_rate_crossing(data)))


# Master list for storing the prosodic features as list of lists.
prosodic_features_list = [log_energy, zero_cross_rate, pitch]
```
